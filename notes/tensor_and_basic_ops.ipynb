### 0.导入包并设置GPU

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL']='2'
os.environ['CUDA_VISIABLE_DEVICES']='O'

import tensorflow as tf 
import numpy as np
phy_gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in phy_gpus:
    tf.config.experimental.set_memory_growth(gpu,True)
```

### 1.创建tensor

```python
a = tf.constant([[1,2],[3,4]])
b = a*a
print(a)
print(b)
print(type(a))
print(type(b))
```

###  2.tensor vs ndarray

```python
c = np.array([[1,2],[3,4]])
d = tf.constant(c)
print(c)
print(d)
print(type(c))
print(type(d))
```

### 3.tensor->ndarray

```python
e = d.numpy()
print(e)
print(type(e))
```

### 4.cpu or gpu

```python
print(d.device)
print(tf.config.list_physical_devices('CPU'))
print(tf.config.list_physical_devices('GPU'))
```

### 5.Variable

```python
f = tf.Variable([[1,2],[3,4]])
print(f)
print(type(f))
```

### 6.修改tensor的值
#### tf.constant不可修改

```python
g = tf.constant(100)
print(g)
g.assign(20)
```

#### tf.Variable可以修改

```python
h = tf.Variable(100)
print(h)
h.assign(50)
print(h)
```
