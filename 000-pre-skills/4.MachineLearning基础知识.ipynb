### 1.知道统计机器学习为什么叫"统计"机器学习
* 能大致说出机器学习与概率统计之间的关系
* 知道频率学派与贝叶斯学派的区别
* 能想象出联合概率分布与条件概率分布的关系

### 2.统计机器学习处理的结构化数据的基本概念
* 属性
* 标签
* 特征
* 实例
* 误差

### 3.统计机器学习的几个分类维度
* 根据处理的数据不同可以分为：分类、回归、标注
* 根据是否有标签数据可以分为：监督、无监督、半监督
* 根据学习的方法不同可以分为：生成方法、判别方法

### 4.能独立判断一个模型是过拟合还是欠拟合
* 什么是交叉验证？
* 什么是训练误差，测试误差，泛化误差？

### 5.了解统计机器学习项目的基本流程
* 明确任务目标及其量化标准
* 训练与测试数据的收集，选择，清洗，分析
* 模型的构建，训练，测试
* 模型的部署，监控，升级
