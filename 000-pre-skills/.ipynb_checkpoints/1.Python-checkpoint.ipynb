### 1.装饰器

能够做到：
* 看得懂装饰器，并且能大概猜到装饰器函数里都写了什么
* 能熟练的使用TensorFlow提供的装饰器工具
* 不要求自己能熟练开发装饰器函数或类似工具


### 2.迭代器

能够做到：

* 知道迭代器的原理，明白迭代器的优势
* 能熟练使用针对TensorFlow开发的迭代器工具
* 特定情况下能根据任务目标，基于迭代器协议自行设计数据准备函数作为数据输入管道


### 3.上下文管理器
能够做到：
* 知道上下文管理器的原理，明白为什么要使用上下文管理器
* 在遇到TensorFlow中的with语句的时候，能看透本质，而不感到陌生

### 4.Python的面向对象编程
能够做到：
* 能熟练在TensorFlow提供的基类上设计定制化继承类
* 熟练掌握**```__init__()```**的各种用法和使用场景
* 知道Python的鸭子类型，掌握几个基本的接口协议

### 5.Python的模块导入机制
能够做到：
* 简单了解Python的导入时与运行时
* 能独立的对代码进行模块的拆分

### 6.Python的pip包管理
能够做到：
* 熟练使用pip进行Python的包管理
